```python
# 设置全局字体为Times New Roman，统一字号
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 16  # 统一基础字号

# 计算每层的统计数据
layer_means = np.mean(data_array, axis=0)
layer_stds = np.std(data_array, axis=0)

# 创建条形图（高度改为3）
fig, ax = plt.subplots(figsize=(12, 4))

# 绘制条形图
bars = ax.bar(range(24), layer_means, yerr=layer_stds, capsize=3, alpha=0.7,
               color='#276CF5', edgecolor='black', linewidth=0.5)

# 根据值的正负给条形图着色
colors = ['crimson' if val > 0 else '#276CF5' for val in layer_means]
for bar, color in zip(bars, colors):
    bar.set_color(color)

ax.axhline(y=0, color='grey', linestyle='-', linewidth=1, alpha=0.3)
ax.set_xlabel('Layer', fontsize=18)
ax.set_ylabel('Head-Averaged Leakage Bias', fontsize=18)
# 不要标题
ax.set_xticks(range(24))
ax.set_xticklabels(range(24), fontsize=16)
ax.tick_params(axis='y', labelsize=16)
ax.grid(axis='y', alpha=0.3)

# 调整y轴范围，给下方留更多空间
y_min = min(layer_means - layer_stds) - 0.1  # 下方多留0.1的空间
y_max = max(layer_means + layer_stds) + 0.05  # 上方留一点空间
ax.set_ylim(y_min, y_max)


# 添加数值标注
for i, (mean, std) in enumerate(zip(layer_means, layer_stds)):
    if abs(mean) > 0.01:  # 只标注绝对值大于0.01的值
        ax.text(i, mean + std + 0.02 if mean > 0 else mean - std - 0.02,
                f'{mean:.2f}', ha='center', va='bottom' if mean > 0 else 'top',
                fontsize=12, rotation=0, fontfamily='Times New Roman')

plt.tight_layout()

# 保存到PDF文件
plt.savefig('./softmaxbiasbarplot.pdf',
            format='pdf',
            dpi=300,
            bbox_inches='tight',
            facecolor='white',
            edgecolor='none')
plt.close()

print("图片已保存为 softmax_offset_barplot.pdf")

# 恢复默认字体设置
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.size'] = 10

# 打印分析
print("\n" + "="*60)
print("层级分析")
print("="*60)
for i in range(24):
    print(f"Layer {i:2d}: mean={layer_means[i]:+.4f}, std={layer_stds[i]:.4f}, "
          f"range=[{data_array[:, i].min():.4f}, {data_array[:, i].max():.4f}]")
```
